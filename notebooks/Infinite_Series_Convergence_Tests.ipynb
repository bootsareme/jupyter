# Infinite Series Convergence Tests

Testing whether an infinite series converges or diverges is crucial in many scientific fields. While no practical application is infinite, the tests described below can show the end behavior of a specific series. Moreover, these tests show if any meaningful results exist before the computation of the sum has started. After all, it would be highly inefficient to compute a series that diverges.

--- 

## $n$th Term Divergence Test

This test only concludes if a series diverges or not. It cannot test for convergence.

- If $\displaystyle \lim_{n\to∞} a_n \neq 0$ or said limit does not exist, the series $\displaystyle \sum_{n=k}^{\infty} a_n$ diverges.

- It cannot be guaranteed that the series $\displaystyle \sum_{n=k}^{\infty} a_n$ converges when $\displaystyle \lim_{n\to∞} a_n = 0$.

One notable series that diverges when $\displaystyle \lim_{n\to∞} a_n = 0$ is the harmonic series:

$$
\displaystyle \sum_{n=1}^{\infty} \frac{1}{n} = 1 + \frac{1}{2} + \frac{1}{3} + \ldots
$$

Taking the limit as $n$ approaches $\infty$,

$$
\displaystyle \lim_{n\to∞} \frac{1}{n} = 0, 
$$

it is apparent that the $n$th Term Test does not test for convergence.

---

## Ratio Test

Suppose $\exists L$ for $\displaystyle L = \lim_{n\to∞} \biggr\rvert \frac{a_{n+1}}{a_n} \biggr\rvert$ where $a_{n+1}$ is the next term of the sequence.

- If $L < 1$, the series $\displaystyle \sum_{n=k}^{\infty} a_n$ is absolutely convergent.
- If $L > 1$, the series $\displaystyle \sum_{n=k}^{\infty} a_n$ diverges.
- If $L = 1$, the Ratio Test is inconclusive.

---

## Root Test

This test involves taking the $n$th root of the absolute value of the $n$th term.

- If $\displaystyle \lim_{n\to∞} \sqrt[n]{|a_n|} < 1$, the series $\displaystyle \sum_{n=k}^{\infty} a_n$ converges absolutely.
- If $\displaystyle \lim_{n\to∞} \sqrt[n]{|a_n|} > 1$, the series $\displaystyle \sum_{n=k}^{\infty} a_n$ diverges.
- If $\displaystyle \lim_{n\to∞} \sqrt[n]{|a_n|} = 1$, the Root Test is inconclusive.

---

## Integral Test

Let $f(x)$ be a continuous decreasing function on the interval $[k, \infty)$ and $f(x) \geq 0, \forall x\geq k$. If $f(n) = a_n$, then:

- $\displaystyle \int_k^∞ f(x) \, \text{d}x  < ∞ \implies \displaystyle \sum_{n=k}^{\infty} a_n$ converges.
- $\displaystyle \int_k^∞ f(x) \, \text{d}x$ diverges $\iff \displaystyle \sum_{n=k}^{\infty} a_n$ diverges.

---

## $p$-Series Test

This is a corollary of the Integral Test described above. Given a series in the form of $\displaystyle \sum_{n=k}^{\infty} \frac{1}{n^p}$, and $k > 0$,

- if $p > 1$, the series converges.
- if $p \leq 1$, the series diverges.

This can be proven given $\displaystyle \int_k^∞ \frac{\text{d}x}{x^p}$ converges when $p > 1$ and diverges when $p \leq 1$.

---

## Alternating Series Test

This is a special test that only works on the alternating series in the form of $\displaystyle \sum_{n=k}^{\infty} (-1)^na_n$ or $\displaystyle \sum_{n=k}^{\infty} (-1)^{n+1}a_n$ where $a_n \geq 0, \forall n$. The alternating series is convergent if both conditions are met:

- $\displaystyle \lim_{n\to∞} a_n = 0$
- $a_{n+1} \leq a_n, \forall n$

The alternating series is divergent if any of the two conditions above fail.

---

## Direct Comparision Test

There are two variants of this test, one for improper integrals and the second for infinite series. For the improper integrals, let $f$ and $g$ be defined on the interval $[k,\infty)$ such that $0 \leq g(x) \leq f(x)$.

- $\displaystyle \int_k^∞ f(x) \, \text{d}x$ converges $ \implies \displaystyle \int_k^∞ g(x) \, \text{d}x$ converges.
- $\displaystyle \int_k^∞ g(x) \, \text{d}x$ diverges $ \implies \displaystyle \int_k^∞ f(x) \, \text{d}x$ diverges.

The second variant applies the improper integral test on infinite series. Given two series $\displaystyle \sum_{n=k}^{\infty} a_n$ and $\displaystyle \sum_{n=k}^{\infty} b_n$ with $b_n \geq a_n \geq 0, \forall n$, the following are true:

- $\displaystyle\sum_{n=k}^{\infty} b_n$ converges $\implies \displaystyle\sum_{n=k}^{\infty} a_n$ converges.
- $\displaystyle\sum_{n=k}^{\infty} a_n$ diverges $\implies \displaystyle\sum_{n=k}^{\infty} b_n$ diverges.

---

## Limit Comparision Test

Given two series $\displaystyle \sum_{n=k}^{\infty} a_n$ and $\displaystyle \sum_{n=k}^{\infty} b_n$ with $b_n > 0, a_n \geq 0, \forall n$, if $0 <\displaystyle \lim_{n\to∞} \frac{a_n}{b_n} < \infty$, then:

- $\displaystyle \sum_{n=k}^{\infty} a_n$ converges $ \iff \displaystyle \sum_{n=k}^{\infty} b_n$ converges.
- $\displaystyle \sum_{n=k}^{\infty} a_n$ diverges $ \iff \displaystyle \sum_{n=k}^{\infty} b_n$ diverges.

---

## Absolute Convergence Test

Given a series $\displaystyle \sum_{n=k}^{\infty} a_n$ and its absolute form $\displaystyle \sum_{n=k}^{\infty} |a_n|$,

- if $\displaystyle \sum_{n=k}^{\infty} a_n$ converges and $\displaystyle \sum_{n=k}^{\infty} |a_n|$ converges, then $\displaystyle \sum_{n=k}^{\infty} a_n$ is **absolutely convergent**.
- if $\displaystyle \sum_{n=k}^{\infty} a_n$ converges but $\displaystyle \sum_{n=k}^{\infty} |a_n|$ diverges, then $\displaystyle \sum_{n=k}^{\infty} a_n$ is **conditionally convergent**.
- if both $\displaystyle \sum_{n=k}^{\infty} a_n$ and $\displaystyle \sum_{n=k}^{\infty} |a_n|$ fail to converge, then $\displaystyle \sum_{n=k}^{\infty} a_n$ diverges.

--- 

## Geometric Series Test

When the series is in the form of a infinite geometric series $\displaystyle \sum_{n=0}^{\infty} ar^n$, then:

- If $|r| > 1$, the geometric series $\displaystyle \sum_{n=0}^{\infty} ar^n$ diverges.
- If $|r| < 1$, the geometric series $\displaystyle \sum_{n=0}^{\infty} ar^n$ converges.

When $|r| = 1$, there are two possibilities:

- If $r = 1$, the geometric series diverges without bounds:

$$
\displaystyle \sum_{n=0}^{\infty} a(1)^n = \displaystyle \sum_{n=0}^{\infty} a = a + a + a + \ldots = ∞
$$

- If $r = -1$, the geometric series diverges through the oscillation of an alternating series:

$$
\displaystyle \sum_{n=0}^{\infty} a(-1)^n = a - a + a - a + a - \ldots
$$

The partial sum for any finite geometric series up to $k$ is as follows:

$$
\sum_{n=0}^{k} ar^n = a + ar + ar^2 + ar^3 + \ldots + ar^k = a(\frac{1 - r^{k+1}}{1 - r}), r \neq 1
$$

As $k$ approaches $∞$, the infinite series converges on

$$
\displaystyle \sum_{n=0}^{\infty} ar^n = \frac{a}{1-r}, |r| < 1.
$$

## Binomial Series Test

The Binomial Series is the Taylor Series for the function $f(x) = (1 + x)^z$ where $z \in \mathbb{C}$. Explicitly, this is defined as:

$$
(1+x)^z = \displaystyle \sum_{n=0}^{\infty} {z \choose n} x^n
= 1 + zx + \frac{z(z-1)}{2!}x^2 + \frac{z(z-1)(z-2)}{3!}x^3 + \ldots,
$$

and the notation for general binomial coefficients is defined as:

$$
{z \choose n} := \frac{z(z-1)(z-2)(z-3)\ldots(z-n+1)}{n!}
$$

Below are the conditions that test for the convergence or divergence of the binomial series:

- If $|x| < 1$, the binomial series $\displaystyle \sum_{n=0}^{\infty} 
{z \choose n} x^n$ converges absolutely $\forall z$.
- If $|x| = 1$, the binomial series $\displaystyle \sum_{n=0}^{\infty} 
{z \choose n} x^n$ converges absolutely $\iff \text{Re}(z) > 0$ or $z = 0$.
- If $x = 1$, the binomial series $\displaystyle \sum_{n=0}^{\infty} 
{z \choose n} x^n$ converges $ \iff \text{Re}(z) > -1$.
- If $x = -1$, the binomial series $\displaystyle \sum_{n=0}^{\infty} 
{z \choose n} x^n$ converges $\iff \text{Re}(z) > 0$ or $z = 0$.
- If $|x| > 1$, the binomial series $\displaystyle \sum_{n=0}^{\infty} 
{z \choose n} x^n$ diverges unless $z \geq 0$.

In particular, if $z \geq 0$, the situation at the boundary of the disk of convergence, $|x| = 1$, is described as follows:

- If $\text{Re}(z) > 0$, the binomial series $\displaystyle \sum_{n=0}^{\infty} 
{z \choose n} x^n$ converges absolutely.
- If $ -1 < \text{Re}(z) \leq 0$, the binomial series $\displaystyle \sum_{n=0}^{\infty} 
{z \choose n} x^n$ converges conditionally if $x \neq -1$ and diverges if $x = -1$.
- If $\text{Re}(z) \leq -1$, the binomial series $\displaystyle \sum_{n=0}^{\infty} 
{z \choose n} x^n$ diverges.
